#### **Data Cleaning**

```python
df.isnull().sum()
```

```python
df['price']=df['price'].fillna(df['price'].mean())
```

```python
df.isnull().sum()
```

##### **1.Removing Nulls**

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

##### **2.Removing Duplicates**

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df.loc[df['rating']=='No rating','rating']=0
```

##### **3.Correcting Data Types**

```python
df['id']=df['id'].astype(object)
df['host_id']=df['host_id'].astype(object)
df['rating'] = pd.to_numeric(df['rating'].astype(str).str.strip(), errors='coerce')
```

### **EDA**

#### **Data Analysis**

##### **1.Univariate Analysis**

```python
df=df[df['price']<1500]
sns.boxplot(data=df,x='price')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.histplot(data=df,x='price',bins=100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=df,x='availability_365',bins=50)
plt.title('availability_365 distribution')
plt.ylabel('Frequency')
plt.show()
```

##### **2.Feature Engineering**

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df['price_by_bed']=df['price']/df['beds']
df.head(5)
```

```python
df.groupby(by='neighbourhood_group')['price_by_bed'].mean()
```

##### **Bi Variable Analysis**

```python
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
plt.title('Price of different rooms of different Locality ')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.title('No of Reviews by price of different Locality')
plt.show()
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.title('Geographical distribution of room_type')
plt.show()
```

```python
##correlation
corr=df[['latitude','longitude','price','minimum_nights','number_of_reviews','availability_365','beds']].corr()
corr
```

```python
plt.title('correlations of different variables')
sns.heatmap(data=corr,annot=True)
plt.figure(figsize=(8,8))
plt.show()
```
